```python
import polars as pl
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Create a sample Polars dataframe
df = pl.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 5, 7, 9]
})

# Plot the dataframe
plt.figure(figsize=(8, 6))
plt.plot(df['x'], df['y'], marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sample Polars DataFrame Plot')
plt.grid(True)
plt.style.use('default')
plt.show()
```
